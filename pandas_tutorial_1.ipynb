```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(0)
mu = 170 #mean
sigma = 6 #stddev
sample = 100
height = np.random.normal(mu, sigma, sample)
weight = (height-100) * np.random.uniform(0.75, 1.25, 100)
```

```python
gym = pd.DataFrame({'height': height, 'weight': weight})
```

```python
gym
```

```python
gym.plot.scatter(x = 'weight', y = 'height')
```

```python
x = gym.weight
y = gym.height
plt.scatter(x,y)
```
